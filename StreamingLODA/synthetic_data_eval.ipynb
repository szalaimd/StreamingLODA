```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from OnlineLODA import OnlineLODA
from sklearn.metrics import roc_auc_score
```

```python
x = np.load('data.npy')
```

```python
memory_res = []
for i in [1,5,10,20,100, 500]:
    b = OnlineLODA(dimension = x.shape[1], X=x, hist_per_batch=1, batch_size=350, memory=i)
    b_preds = np.hstack(b.predictions)
    memory_res.append(b_preds)
```

```python
hpb_res = []
for i in [1,5,10]:
    b = OnlineLODA(dimension = x.shape[1], X=x, hist_per_batch=i, batch_size=350, memory=5)
    b_preds = np.hstack(b.predictions)
    hpb_res.append(b_preds)
```

```python
plt.figure(figsize=(20, 5))

plt.subplot(3, 1, 1)  # 3 rows, 1 column, subplot 1
plt.plot(hpb_res[0])
plt.title('$k_b$ = 1', size=30)
plt.xlim(-10000, 710000)
plt.tick_params(axis='both', which='major', labelsize=20)  

plt.subplot(3, 1, 2)  # 3 rows, 1 column, subplot 2
plt.plot(hpb_res[1])
plt.title('$k_b$ = 5', size=30)
plt.xlim(-10000, 710000)
plt.tick_params(axis='both', which='major', labelsize=20)  

plt.subplot(3, 1, 3)  # 3 rows, 1 column, subplot 3
plt.plot(hpb_res[2])
plt.title('$k_b$ = 10', size=30)
plt.xlim(-10000, 710000)
plt.tick_params(axis='both', which='major', labelsize=20)  

plt.tight_layout()  # Adjust spacing
plt.savefig('hpb_synth2.pdf')
```
