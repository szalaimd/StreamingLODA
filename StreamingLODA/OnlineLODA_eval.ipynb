```python
dta_label = pd.read_csv('../datasets/sensor.csv', index_col=0)['machine_status']
```

```python
def transformer(x):
    if(x=='NORMAL'):
        return 0
    else:
        return 1
```

```python
dta_label_bin = dta_label.apply(transformer)
```

```python
b = OnlineLODA(dimension = dta.shape[1], X=dta, batch_size=1440)
```

```python
plt.figure(figsize=(30,10))
plt.scatter(np.arange(res.shape[0])[dta_label_bin==0], res[dta_label_bin==0], c='blue', s=0.1)
plt.scatter(np.arange(res.shape[0])[dta_label_bin==1], res[dta_label_bin==1], c='red', s=0.1)
plt.show()
```

```python
datetime_series = pd.Series(
    pd.date_range("2000-01-01", periods=len(dta_label_bin), freq="min")
)#.dt.time
datetime_series.dt.dayofyear
```

```python
import matplotlib.dates as mdates


def anosc_plot(preds, filename = None):
    fig = plt.figure(figsize=(15,5))
    ax = fig.add_subplot(111)
    print(preds.shape)
    ax.scatter(datetime_series[dta_label_bin==0], preds[dta_label_bin==0], color="blue", s=0.2, rasterized=True)
    ax.scatter(datetime_series[dta_label_bin==1], preds[dta_label_bin==1], color="red", s=0.2, rasterized=True)
    
    # ax.xaxis.set_major_locator(mdates.DayLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%j'))
    ax.legend(['Normal', 'Anomaly/Recovering'])
    
    # ax.vlines(mixed_dates, 0,35, linewidth=0.2, color="black")
    plt.tight_layout()
    
    if filename is not None:
        plt.savefig(filename, dpi=300)
    plt.show()
```

```python
b_results = pd.DataFrame(columns=np.arange(20))

for i in np.arange(20):
    b = OnlineLODA(dimension = dta.shape[1], X=dta, batch_size=1440)
    b_preds = np.hstack(b.predictions)
    b_results[i] = b_preds
    
    print(f"round {i+1} is done")
```

```python
anosc_plot(b_results.mean(axis=1), filename=None)
```
